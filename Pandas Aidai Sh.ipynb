Which column has the highest percentage of null values? What is the percentage?

```python
null_percentages = dataframe.isnull().sum() / len(dataframe) * 100
max_null_percent = null_percentages.max()
print(f'The highest percentage is \'director\', {max_null_percent:.2f}%')
```

The way we handle null values depends on how valuable the columns and rows are to us, and the datatype. 

Since the director column could be important to our downstream analysis, we want to keep the column. However, we need to deal with the null values. In this case, since it is a string column, let us replace the null values with the empty string. Replace the null values in the original dataset.

```python
dataframe['director'].fillna('', inplace=True)
dataframe
```

### Exploratory Data Analysis

What is the average (rounded) release year for the Netflix titles listed?

```python
print(round(dataframe['release_year'].mean()))
```

Which country has the highest number of movie titles? What is that number? What is the percentage of all movies?

```python
movies_by_country = dataframe.groupby('country')['title'].count()

max_movies_country = movies_by_country.idxmax()
max_movies_count = movies_by_country[max_movies_country]

total_movies = dataframe['title'].count()
max_movies_percent = (max_movies_count / total_movies) * 100

print(f'The country with the highest number of movie titles is {max_movies_country} with {max_movies_count} titles.')
print(f'This represents {max_movies_percent:.2f}% of all {total_movies} movies in the dataset.')
```
